# Pokemon Data Analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Data Overview

```python
file_name = "pokemon.csv"
data = pd.read_csv(file_name)
```

#### Column Labels

```python
column_names = data.columns
", ".join(column_names)
```

```python
print(data["base_egg_steps"])
```

#### Dataset Snapshot

```python
#set pokedex number as index
data.set_index("pokedex_number")
```

## Summary

#### Number of Pokemon in Dataset

```python
data["name"].count()
```

#### Pokemon Types

```python
types = list(set(data["type1"]))
types.sort()
", ".join(types)
```

#### Ratio of Pokemon with a Single Type Versus a Dual Type

```python
single_type = data[data["type2"].isna()]
dual_type = data[data["type2"].notna()]

single_dual = pd.DataFrame({"Count": [single_type["name"].count(), dual_type["name"].count()]}, index=["Single Type", "Dual Type"])
single_dual.plot.pie(y="Count")
```

#### Ratio of Different Types in Single Type Pokemon

```python
single_grouped = single_type.groupby("type1").size()
single_grouped.plot.pie(y="type1", figsize=(10, 10))
```

#### Weight Distribution by Single Type

```python
single_type_weight = single_type[["weight_kg", "type1"]].groupby("type1")
single_type_weight = single_type_weight.boxplot(figsize=(10,20), sharey=False)
```

#### Height Distribution by Single Type

```python
single_type_height = single_type[["height_m", "type1"]].groupby("type1")
single_type_height = single_type_height.boxplot(figsize=(10,20), sharey=False)
```

## Correlation

#### Height versus Weight

```python
data[["weight_kg", "height_m"]].corr()
```

```python
x = data["weight_kg"]
y = data["height_m"]
plt.xlabel("Weight (kg)")
plt.ylabel("Height (m)")
plt.scatter(x, y)
```

#### Attack versus Defense

```python
data[["attack", "defense"]].corr()
```

```python
x = data["attack"]
y = data["defense"]
plt.xlabel("Attack")
plt.ylabel("Defense")
plt.scatter(x, y)
```

Defense versus Weight

```python
data[["defense", "weight_kg"]].corr()
```

```python
x = data["defense"]
y = data["weight_kg"]
plt.xlabel("Defense")
plt.ylabel("Weight (kg)")
plt.scatter(x, y)
```
